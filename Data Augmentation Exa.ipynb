# Case-1 (Lion image)

```python
#Load an image
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
import numpy as np

#Loading an image
image = load_img("Lion.jpg" , target_size=(100,100))
plt.imshow(image)
#Convert the image to numpy array
image = img_to_array(image)

#Whenever we are planning to do preprocessing or training, it is expected that there must exists 4 dimension
# (numImages, height(y), width(x), channel)
image = np.expand_dims(image, axis=0)
image.shape
```

```python
# Data Augmentation

from tensorflow.keras.preprocessing.image import ImageDataGenerator

augmenter = ImageDataGenerator(rotation_range=30,  #Rotation
                              zoom_range=0.2,      #Zooming
                              horizontal_flip=True, #Horizontal Flip
                              fill_mode='nearest')
```

```python
imageGen = augmenter.flow(image,
                         batch_size=1,
                         save_to_dir="Lion_Directory", #Ensure the directory exists else prog will throw error
                         save_prefix="lion-",
                         save_format="jpg")


counter = 0

for generatedImage in imageGen:
    
    counter += 1 
    
    if counter == 100: #The loop breaks after it generates 100 images
        break
```
